```python
import re # 
def remove_links(description):
    description = re.sub('@[^\s]+',' ',str(description))
    description = re.sub('http[^\s]+',' ',str(description))
    description = re.sub('<[^\s]+',' ',str(description))
        
    return description
df['description'] = df['description'].apply(remove_links)
```

```python
df.head()
```

```python
# Convert all messages to lower case
df['description'] = df['description'].str.lower()

# Replace URLs with 'webaddress'
df['description'] = df['description'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$','webaddress')

df['description'] = df['description'].str.replace(r'£|\$', 'dollars')# Replace money symbols with 'moneysymb' 

# Replacing 10 digit phone numbers with 'phonenumber'
df['description'] = df['description'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$','phonenumber')

df['description'] = df['description'].str.replace(r'\d+(\.\d+)?','num') # Replace numbers with 'num'

df['description'] = df['description'].str.replace(r'[^\w\d\s]',',') #removing punctuations

df['description'] = df['description'].str.replace(r'[\_]',' ') #removing underscore characters

df['description'] = df['description'].str.replace(r'\s+[a-zA-Z]\s+', ' ') #removing single characters

df['description'] = df['description'].str.replace(r'\s+', ' ') #removing whitespace between terms with a single space

df['description'] = df['description'].str.replace(r'^\s+|\s+?$', ' ') #removing leading and trailing whitespace
```

```python
KAR = ['bengaluru','bangalore','benguluru','bengaluru (garvebhavipalya)','bangalote','bengalure','bangalore rural',
      'electronic city','banagalore','garvebhavi palya','bidadi','basavanapura','kurnool']

TN = ['peelamedu','kalapatti','coimbatore','tirunelveli','saravanampatti','chennai',
        'oragadam','guduvancheri','mudichur','kancheepuram','vengadasalapuram']

MAHA = ['nashik','nanekarwadi','pune','mumbai','nasik']

AP = ['hyderabad','madhapur']

GUJ = ['iyava','bhavnagar']

NORTH = ['new delhi','dehradun','rewari','kudlu','pinjore,','kurali']
# Grouping into states to avoid single entry and bias entry
```

```python
for i in KAR:
    df.loc[df["value"] == i , "value"] = "karnataka"
for i in TN:
    df.loc[df["value"] == i , "value"] = "tamil nadu"
for i in MAHA:
    df.loc[df["value"] == i , "value"] = "maharashtra"
for i in AP:
    df.loc[df["value"] == i , "value"] = "andhra pradesh"
for i in GUJ:
    df.loc[df["value"] == i , "value"] = "gujrat"
for i in NORTH:
    df.loc[df["value"] == i , "value"] = "north india"
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords,wordnet

cv = CountVectorizer(stop_words = 'english')
words = cv.fit_transform(df.description)

sum_words = words.sum(axis=0)

words_freq = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)

frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])

frequency.head(20).plot(x='word', y='freq', kind='bar', figsize=(15, 7), color = 'green')
plt.title("Most Frequently Occuring Words - Top 20")
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english') + ['u','m', 'ü', 'ur', '4', '2', 'im', 'dont'])
df['description'] = df['description'].apply(lambda x: ' '.join(term for term in x.split() if term not in stop_words))
```

```python
lem=WordNetLemmatizer()
df['description'] = df['description'].apply(lambda x: ' '.join(lem.lemmatize(t) for t in x.split()))
```
